```python
import json
import pandas as pd
import folium
from folium import plugins
```

```python
geo = json.load(open('map_people.geojson', encoding='utf-8'))
geo
```

```python
df = pd.read_csv('행정구역_읍면동_별_5세별_주민등록인구_2011년__20220506233554.csv',encoding='cp949')
df
```

```python
m = folium.Map(location=[36.1,128.1],zoom_start=9.5)
m.choropleth(geo_data=geo, data=df, columns=['name','people'],key_on='properties.name',
             fill_color='PuRd',legend_name='인구 수 (명)')


m
```

```python
df2 = pd.read_csv('위도경도.csv')

df2['주소'][0]
```

```python
from folium.plugins import MarkerCluster

marker_cluster = MarkerCluster().add_to(m)


for i in range(0,8):
    folium.Marker(location=[df2['위도'][i], df2['경도'][i]],
                  popup=df2['주소'][i]).add_to(marker_cluster)

folium.Marker(location=[36.16488389999973,128.13139059999966],
              popup='김천시 물류센터', icon=folium.Icon(color='red')).add_to(m)
m
```

```python
m.save('people_center+smartcenter.html')
```
